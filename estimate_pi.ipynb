```python
# import packages
import random
import matplotlib.pyplot as plt
import turtle
import math
```

```python
# to visualize random points
my_pen = turtle.Turtle()
my_pen.hideturtle()
my_pen.speed(0)
```

```python
# draw a square
square_length = 200

my_pen.up()

# set initial point
my_pen.setposition(-square_length/2, -square_length/2)
my_pen.down()

# draw right side
my_pen.fd(square_length)
my_pen.left(90)

# draw top side
my_pen.fd(square_length)
my_pen.left(90)

# draw left side
my_pen.fd(square_length)
my_pen.left(90)

# draw bottom side
my_pen.fd(square_length)
my_pen.left(90)
```

```python
# drow a circle
radius = square_length/2

my_pen.up()
my_pen.setposition(0, -radius)
my_pen.down()
my_pen.circle(radius)
```

```python
# count points inside and outside the circle
in_circle_count = 0
out_circle_count = 0

# list to store values of pi
pi_val_list = []
```

```python
num_iterations = 5
for i in range(num_iterations):
    
    # number of samples for each iteration
    num_samples = 100
    
    for j in range(num_samples):
        # generate random point
        x = random.randrange(-square_length/2, square_length/2)
        y = random.randrange(-square_length/2, square_length/2)

        # check if point is out of circle
        if (x**2 + y**2 > radius**2):
            my_pen.color('blue')
            my_pen.up()
            my_pen.goto(x, y)
            my_pen.down()
            my_pen.dot()
            out_circle_count = out_circle_count + 1

        # in 
        else:
            my_pen.color("red")
            my_pen.up()
            my_pen.goto(x,y)
            my_pen.down()
            my_pen.dot()
            in_circle_count = in_circle_count + 1

        # calculate value of pi
        pi = 4.0 * in_circle_count / (in_circle_count + out_circle_count)

        #Append the values of PI in list :
        pi_val_list.append(pi)

        # calculate error of each value of pi
        pi_error_list = [abs(math.pi - pi) for pi in pi_val_list]

    # print value of pi for each iteration
    print(pi_val_list[-1])
```

```python
# plot values of pi
plt.axhline(y=math.pi, color='g', linestyle='-')
plt.plot(pi_val_list)
plt.xlabel("Iterations")
plt.ylabel("Value of PI")
plt.show()
```

```python
# plot error in the estimate values
plt.axhline(y=0.0, color='g', linestyle='-')
plt.plot(pi_error_list)
plt.xlabel("Iterations")
plt.ylabel("Error")
plt.show()
```
