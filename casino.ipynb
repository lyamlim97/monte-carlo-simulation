```python
# import packages
import random
import matplotlib.pyplot as plt
```

```python
# user decide to bet on odd or even
choice = 'even'

# function for even, returns if win or lose
# True = win, False = lose
if choice=="even":
    def pick_chip():
        # random nunmber betwwen 1 and 100
        chip = random.randint(1, 100)
               
        # note that 10 is not considered as an even number.
        if chip%2!=0 or chip==10:
            return False
        elif chip%2==0:
            return True

# function for odd
if choice=="odd":
    def pick_chip():
        # random nunmber betwwen 1 and 100
        chip = random.randint(1, 100)
               
        # note that 11 is not considered as an odd number.
        if chip%2==0 or chip==11:
            return False
        elif chip%2!=0:
            return True
```

```python
# main function
def play(initial_money, bet_money, num_bets):

    bet_count_list = []
    money_track_list = []
    
    for bet_count in range(1, num_bets+1):
        # if win
        if pick_chip():
            # add to our money
            initial_money = initial_money + bet_money
            # appent play number
            bet_count_list.append(bet_count)
            # append to money tracking list
            money_track_list.append(initial_money)
        
        # if lose
        else:
            # remove from our money
            initial_money = initial_money - bet_money
            # appent play number
            bet_count_list.append(bet_count)
            # append to money tracking list
            money_track_list.append(initial_money)
    
    # plot the data
    plt.ylabel('Player Money in $')
    plt.xlabel('Number of bets')
    plt.plot(bet_count_list, money_track_list)

    #Final value after all the iterations :
    final_fund.append(money_track_list[-1])
    return(final_fund)
```

```python
# list for final funds
final_fund = []

num_players = 10
initial_money = 10000
bet_money =  100
num_bets = 50
    
for i in range(num_players):
    ending_fund = play(initial_money, bet_money, num_bets)
    
# print starting and ending amount of money for average player 
print("All players started with $", initial_money )
print("The average player will be left with $", sum(ending_fund)/len(ending_fund))
```
