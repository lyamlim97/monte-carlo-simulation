```python
# import packages
import random
import math
import matplotlib.pyplot as plt
```

```python
# estimate value of pi
def monte_carlo(num_iterations, num_samples, needle_length, box_width):
    # list for pi values
    pi_val_list = []
    
    # horizontal line for actual value of pi
    plt.axhline(y=math.pi, color='r', linestyle='-')
    
    # for each iteration
    for i in range(num_iterations):
        # set initial number of hits as 0
        num_hits = 0
        
        # for all samples, i.e. needles in this iteration
        for j in range(num_samples):
            # find distance from the nearest vertical line
            # min = 0
            # max = box_width/2
            x = random.uniform(0, box_width/2)
            
            #The theta value will be from 0 to pi/2 :
            theta = random.uniform(0, math.pi/2)
            
            # check if needle crosses line
            x_tip  = x - (needle_length/2.0)*math.cos(theta)  
            if x_tip < 0 :
                num_hits += 1
                
        # applying formula
        numerator = 2.0 * needle_length * num_samples
        denominator = box_width * num_hits
       
        # append final value of pi
        pi_val_list.append((numerator/denominator))
    
    # print final value of pi of each iteration
    print(pi_val_list[-1])
    
    # plot graph
    plt.plot(pi_val_list)
```

```python
num_iterations = 500
num_samples = 10000
needle_length = 2
box_width = 2

monte_carlo(num_iterations, num_samples, needle_length, box_width)
```
